```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1
lb = q1 - (1.5*iqr)
ub = q3 + (1.5*iqr)
print(lb,ub)
```

```python
df = df[(df>lb) & (df<ub)]
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,:-1]
X.head()
```

```python
y = df.iloc[:,-1].values
y
```

```python
y.reshape(-1,1)
```

```python
X.head()
```

Splitting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
X_train.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
ss_x = StandardScaler()
ss_y = StandardScaler()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
reg.predict(X).reshape(-1,1)
```

```python
reg.coef_
```

```python
reg.score(X_test,y_test)
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree =2)
X_poly = poly.fit_transform(X_train)
X_poly_test = poly.fit_transform(X_test)
polyreg = LinearRegression()
polyreg.fit(X_poly,y_train)
```

```python
polyreg.score(X_poly_test,y_test)
```

```python
poly3 = PolynomialFeatures(degree =3)
X_poly3 = poly3.fit_transform(X_train)
X_test_poly3 = poly3.fit_transform(X_test)

polyreg3 = LinearRegression()
polyreg3.fit(X_poly3,y_train)
```

```python
polyreg3.score(X_test_poly3,y_test)
```

## Support Vector Machines

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train,y_train)
```

```python
svr.score(X_test,y_test)
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state = 0)
dtr.fit(X_train,y_train)
```

```python
dtr.score(X_test,y_test)
```

## Lasso-Ridge Regressor

```python
from sklearn.linear_model import Ridge
ridgeR = Ridge(alpha = 3)
ridgeR.fit(X_train,y_train)
```

```python
ridgeR.score(X_test,y_test)
```
