```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan,strategy = 'mean')
df['normalized-losses'] = imp.fit_transform(df.filter(['normalized-losses']))
df.info()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
categ = ['fuel-type','aspiration','num-of-doors','engine-location']
for feat in categ:
    df[feat] = le.fit_transform(df[feat])
df
```

```python
fl = ['make','body-style','drive-wheels','fuel-system','engine-type','num-of-cylinders']
for feat in fl:
    dummies = pd.get_dummies(df[feat],drop_first = True)
    df.drop([feat],axis = 1,inplace = True)
    df = pd.concat([df,dummies],axis = 1)
df
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree =2)
X_poly = poly.fit_transform(X)

polyreg = LinearRegression()
polyreg.fit(X_poly,y)
```

```python
X.shape
```

```python
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(polyreg.predict(X_poly),columns=['MilesMM']).MilesMM,label = "Predicted")
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(y,columns=['MilesMM']).MilesMM,label = "actual")
plt.title("2nd degree")
plt.legend()
```
