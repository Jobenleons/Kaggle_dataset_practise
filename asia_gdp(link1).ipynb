```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("Datasets/link1/asia_gdp.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Indicator Name'].value_counts()
```

```python
df= df.drop(['Indicator Name'],axis=1)
```

```python
df
```

```python
df_pivot = df.pivot_table(columns="Country Name" )
df_pivot
```

```python
df_pivot['year'] = df_pivot.index
df_pivot
```

```python
df_pivot.columns
```

```python
dfm = df_pivot.melt('year',var_name='country', value_name='gdp')
dfm
```

```python
dfm['year'] = pd.to_numeric(dfm['year'])
```

```python
sns.lineplot(x = "year",y = "gdp",data = dfm)
```

Hence there is no null element.

```python
y = dfm.iloc[:,-1].values
X = dfm.drop(['gdp'], axis = 1)
X
```

```python
y= y.reshape(-1,1)
```

```python
y
```

```python
X.head()
```

```python
dummies_country = pd.get_dummies(X.country,drop_first = True)
```

```python
dummies_country
```

```python
X = pd.concat([X.filter(['year']),dummies_country],axis = 1)
```

```python
X
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x = X["year"],y = pd.DataFrame(y ,columns = ["gdp"]).gdp, label = "actual")
sns.lineplot(x = X["year"],y = pd.DataFrame(reg.predict(X) ,columns = ["gdp"]).gdp,label = "predicted")
plt.legend()
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree =3)
X_poly = poly.fit_transform(X)

polyreg = LinearRegression()
polyreg.fit(X_poly,y)
```

```python
sns.lineplot(x = X["year"],y = pd.DataFrame(y ,columns = ["gdp"]).gdp, label = "actual")
sns.lineplot(x = X["year"],y = pd.DataFrame(polyreg.predict(X_poly) ,columns = ["gdp"]).gdp,label = "predicted")
plt.legend()
```

## Support Vector Machines

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X,y)
```

```python
sns.lineplot(x = X["year"],y = pd.DataFrame(y ,columns = ["gdp"]).gdp, label = "actual")
sns.lineplot(x = X["year"],y = pd.DataFrame(svr.predict(X) ,columns = ["gdp"]).gdp,label = "predicted")
plt.legend()
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state = 0)
dtr.fit(X,y)
```

```python
sns.lineplot(x = X["year"],y = pd.DataFrame(y ,columns = ["gdp"]).gdp, label = "actual")
sns.lineplot(x = X["year"],y = pd.DataFrame(dtr.predict(X) ,columns = ["gdp"]).gdp,label = "predicted")
plt.legend()
```

## Lasso-Ridge Regressor

```python
from sklearn.linear_model import Ridge
ridgeR = Ridge(alpha = 5)
ridgeR.fit(X,y)
```

```python
sns.lineplot(x = X["year"],y = pd.DataFrame(y ,columns = ["gdp"]).gdp, label = "actual")
sns.lineplot(x = X["year"],y = pd.DataFrame(ridgeR.predict(X) ,columns = ["gdp"]).gdp,label = "predicted")
plt.legend()
```
