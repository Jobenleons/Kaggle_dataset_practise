```python
df_wod = df.drop(['Month'],axis =1)
df_wod
```

```python
X= df_wod.drop(['Sales'],axis =1)
X
```

```python
y = df_wod.iloc[:,0].values
y
```

```python
y= y.reshape(-1,1)
```

```python
df_wod.corr()
```

```python
X.head()
```

```python
X.count()
```

```python
X1 = pd.concat([df['month_parsed'].dt.month,df['month_parsed'].dt.year],axis = 1)
X1.columns = ['month','year']
X1
```

```python
dummies_month = pd.get_dummies(X1.month,drop_first=True)
dummies_month
dummies_year = pd.get_dummies(X1.year,drop_first=True)
dummies_year
```

```python
X1_ohe = pd.concat([dummies_month,dummies_year],axis = 1)
X1_ohe
```

Splitting

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X1_ohe,y)
```

```python
import matplotlib.pyplot as plt
```

```python
reg.predict(X1_ohe)
```

```python
sns.lineplot(x=X['month_parsed'].dt.year,y=pd.DataFrame(y,columns=['sales']).sales).
sns.lineplot(x=X['month_parsed'].dt.year,y=pd.DataFrame(reg.predict(X1_ohe).reshape(-1,1),columns=['sales']).sales)
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree =3)
X_poly = poly.fit_transform(X)

polyreg = LinearRegression()
polyreg.fit(X_poly,y)
```

```python
X.shape
```

```python
sns.lineplot(x=X['month_parsed'].dt.year,y=pd.DataFrame(y,columns=['sales']).sales)
sns.lineplot(x=X['month_parsed'].dt.year,y=pd.DataFrame(polyreg.predict(X_poly).reshape(-1,1),columns=['sales']).sales)
```

## Support Vector Machines

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X,y)
```

```python
svr.predict(X)
```

```python
sns.lineplot(x=X['month_parsed'].dt.year,y=pd.DataFrame(y,columns=['sales']).sales)
sns.lineplot(x=X['month_parsed'].dt.year,y=pd.DataFrame(svr.predict(X).reshape(-1,1),columns=['sales']).sales)
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state = 0)
dtr.fit(X,y)
```

```python
sns.lineplot(x=X['month_parsed'].dt.year,y=pd.DataFrame(y,columns=['sales']).sales)
sns.lineplot(x=X['month_parsed'].dt.year,y=pd.DataFrame(dtr.predict(X).reshape(-1,1),columns=['sales']).sales)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
X_train.isnull().sum()
```
