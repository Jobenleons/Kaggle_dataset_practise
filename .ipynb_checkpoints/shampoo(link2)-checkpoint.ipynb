```python
df['day'] = df.date_parsed.dt.day
df
```

```python
df['month'] = df.date_parsed.dt.month
df
```

```python
df_wod = df.drop(['date','date_parsed'],axis =1)
df_wod
```

```python
dummies_day = pd.get_dummies(df_wod.day,drop_first = True)
dummies_day
```

```python
dummies_month = pd.get_dummies(df_wod.month,drop_first = True)
dummies_month.rename(columns ={2:'feb',3:'mar',4:'apr',5:'may',6:'jun',7:'jul',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'}, inplace = True)
dummies_month
```

```python
X = pd.concat([dummies_month,dummies_day], axis = 1)
```

```python
X
```

```python
y = df_wod.iloc[:,0].values
y
```

```python
y.reshape(-1,1)
```

```python
X.corr()
```

```python
X.head()
```

Splitting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
X_train.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
ss_x = StandardScaler()
ss_y = StandardScaler()
```

```python
X.columns = X.columns.astype(str)
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)
```

```python
y=y.reshape(-1,1)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
import matplotlib.pyplot as plt
```

```python
reg.predict(X).reshape(-1,1)
```

```python
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(reg.predict(X),columns=['births']).births)
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(y,columns=['births']).births)
```

```python
plt.title("Regression line vs actual points",fontsize = 25,color='green')
plt.ylabel("births")
plt.xlabel("time")
plt.scatter(df['date_parsed'],pd.DataFrame(y,columns=['births'])['births'],s = 0.1,c = ['#1f77b4'])
plt.plot(df['date_parsed'],pd.DataFrame(reg.predict(X),columns=['births']).births)
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree =2)
X_poly = poly.fit_transform(X)

polyreg = LinearRegression()
polyreg.fit(X_poly,y)
```

```python
X.shape
```

```python
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(polyreg.predict(X_poly),columns=['births']).births)
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(y,columns=['births']).births)
plt.title("2nd degree")
```

```python
poly3 = PolynomialFeatures(degree =3)
X_poly3 = poly3.fit_transform(X)
polyreg3 = LinearRegression()
polyreg3.fit(X_poly3,y)
```

```python
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(polyreg3.predict(X_poly3),columns=['temp']).temp)
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(y,columns=['temp']).temp)
plt.title("3rd degree")
```

## Support Vector Machines

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X,y)
```

```python
svr.predict(X)
```

```python
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(svr.predict(X).reshape(-1, 1),columns=['births']).births)
sns.lineplot(x=df_wod['month'],y=pd.DataFrame(y,columns=['births']).births)
```
