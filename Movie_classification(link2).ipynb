```python
df.isnull().sum()
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1
lb = q1 - (1.5*iqr)
ub = q3 + (1.5*iqr)
print(lb,ub)
```

```python
df1 = df.filter(['3D_available','Genre','Start_Tech_Oscar'])
df1
```

```python
df = df[(df>lb) & (df<ub)].drop(['3D_available','Genre','Start_Tech_Oscar'],axis = 1)
df
```

```python
df = pd.concat([df,df1],axis = 1)
df.dropna(inplace = True)
df
```

```python
df.corr()
```

```python
df.isnull().sum()
```

Hence there is no null element.

```python
y = df.iloc[:,-1].values
X = df.drop(['Start_Tech_Oscar'], axis = 1)
X
```

```python
y= y.reshape(-1,1)
```

```python
y
```

```python
X.head()
```

```python
X.Genre.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le1 = LabelEncoder()
X.iloc[:,-2] = le1.fit_transform(X.iloc[:,-2])
```

```python
X.head()
```

```python
d1 = pd.get_dummies(X.Genre,drop_first = True)
d1
```

```python
X = X.drop(['Genre'],axis = 1)
X= pd.concat([X,d1],axis = 1)
X
```

Splitting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
X_train.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
ss_x = StandardScaler()
ss_y = StandardScaler()
X_train.isnull().sum()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(ss_x.fit_transform(X_train),ss_y.fit_transform(y_train))
```

```python
X_test
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
plt.plot(np.linspace(0,5,76),y_test)
plt.plot(np.linspace(0,5,76),ss_y.inverse_transform(reg.predict(ss_x.fit_transform(X_test))))
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree =1)
X_poly_train = poly.fit_transform(X_train)
X_poly_test = poly.fit_transform(X_test)
polyreg = LinearRegression()
polyreg.fit(X_poly_train,y_train)
```

```python
plt.plot(np.linspace(0,5,76),y_test)
plt.plot(np.linspace(0,5,76),polyreg.predict(X_poly_test))
```

## Support Vector Machines

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(ss_x.fit_transform(X_train),ss_y.fit_transform(y_train))
```

```python
svr.predict(ss_x.fit_transform(X_test))
```

```python
import seaborn as sns
sns.histplot(y_test)
sns.histplot(ss_y.inverse_transform(svr.predict(ss_x.fit_transform(X_test)).reshape(-1,1)))
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state = 0)
dtr.fit(ss_x.fit_transform(X_train),ss_y.fit_transform(y_train))
```

```python
pd.DataFrame(y_test,columns =['Start_Tech_Oscar']).value_counts()
```

## Lasso-Ridge Regressor

```python
from sklearn.linear_model import Ridge
ridgeR = Ridge(alpha = 1)
ridgeR.fit(ss_x.fit_transform(X_train),ss_y.fit_transform(y_train))
```

```python
plt.plot(np.linspace(0,5,76),y_test)
plt.plot(np.linspace(0,5,76),ss_y.inverse_transform(ridgeR.predict(ss_x.fit_transform(X_test)).reshape(-1,1)))
```
